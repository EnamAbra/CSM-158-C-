# **Question 1**



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris
iris = load_iris()
df= pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

```python
df.shape
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='.')
#
```

```python
x= df.drop(['target'], axis='columns')
y = df.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("F1 Score:", f1)
```

Accuracy  of 1.0 indicates that 100% of test samples were classified correctly.

Precision of 1.0 also indicates that
every instance predicted as positive was truly positive

F1 Score of 1.0 means there is a prfect balance of precision and recall

This means there are no false positives and no false negatives.

## **Question 2**

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df= pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
x=diabetes.data[:,2].reshape(-1,1)
y=diabetes.target
#this selects the independent variable which is BMI since it has index of 2
#reshape is used because sklearn expexts a 2D array for x
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#data is plit into 80% training and 20% testing
```

```python
x_train
```

```python

linearmodel = LinearRegression()
#linear regression model is initialised
```

```python
linearmodel.fit(x_train,y_train)
#model is trained
```

```python
linearmodel.predict(x_test)
#model i used to predict test set
```

```python
y_test
```

```python
print(f"x shape: {x.shape}")
print(f"y shape: {y.shape}")
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(f"Mean Squared Error :{mean_squared_error(y_test,linearmodel.predict(x_test))})")
rmse = np.sqrt(mean_squared_error(y_test, linearmodel.predict(x_test)))
print(f"Root Mean Squared Error: {rmse}")
```

The Mean Squared Error of 4061.83 and Root Mean Squared Error of 63.73 indicate that the model’s predictions deviate substantially from the actual diabetes progression values.
This suggests that while the model learns some patterns, its predictive accuracy is limited and could be improved.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.scatter(x_test,y_test,color='red',marker='+')
plt.plot(x_test,linearmodel.predict(x_test),color='blue',label='Regression Line')
plt.legend()
plt.show()
```
